```python
def recombination_leakage(idx):
    I_0, I, n, k, T, V, q = sp.symbols("I_0, I, n, k, T, V, q")
    base_eq = sp.Eq(I, I_0*(sp.exp((q*V)/(k*T)) - 1))

    D_p, n_i, L_p, N_D = sp.symbols("D_p, n_i, L_p, N_D")
    I0_eq = sp.Eq(I_0, ((q*D_p*n_i**2)/(L_p*N_D)))

    mob, kb, Eg = sp.symbols("m^{*}, k_B, E_g") 
    ni_eq = sp.Eq(n_i, 2*((mob*kb*T)/(2*sp.pi*hbar**2))**(sp.UnevaluatedExpr(3/2))*sp.exp(-Eg/(2*kb*T)), evaluate=False)
    
    equations = [base_eq, I0_eq, ni_eq]
    display(equations[idx])
    return equations[idx]

for i in range(3):
    recombination_leakage(i)
```

```python
def diffusion_calc(T, K, E):
    return np.array([K*t**3*np.exp(-E/t) for t in T])

def shockley_calc(T, K, E):
    return np.array([K*t**(3/2)*np.exp(-E/(2*t)) for t in T])

temp = np.linspace((273-30), (273+120), 100)
diff = diffusion_calc(temp, 1.6e-19*5e17, 1.19*1.6e-19/1.3e-23)
shock = shockley_calc(temp, 1.6e-19*1e12, 1.19*1.6e-19/1.3e-23)
plt.semilogy([1/t for t in temp], diff)
plt.semilogy([1/t for t in temp], shock)
plt.semilogy([1/t for t in temp], diff+shock)
```
